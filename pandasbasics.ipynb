```python
#Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.
# In Pandas, a DataFrame is a two-dimensional, tabular data structure that consists of rows and columns,
# where each column can be of a different data type. It is a fundamental object in Pandas and is similar
# to a spreadsheet or a SQL table. The DataFrame allows you to store and manipulate structured data easily,
# making it one of the most commonly used data structures in data analysis.

# On the other hand, a Series is a one-dimensional array-like object in Pandas, similar to a column
# in a DataFrame. It is designed to represent a single column or row of data. A DataFrame can be
# thought of as a collection of Series, where each Series represents a column.
#Examples
data1=[1,2,3,4,4,5]
ds=pd.Series(data1)
df1=pd.DataFrame({'Numbers':data1})
print(ds)
print(df1)
```

#Q6. Which of the following is mutable in nature Series, DataFrame, Panel?

In Pandas, both `Series` and `DataFrame` are mutable, meaning that you can modify their values, add or remove elements, and perform various other operations that change their content. A `Series` is mutable in the sense that you can modify individual elements or extend the series with new elements. Similarly, a `DataFrame` is mutable, allowing you to modify data within columns, add or drop columns, and make other changes to the structure.

On the other hand, the `Panel` data structure in Pandas has been deprecated since version 0.25.0, and it's recommended to use `MultiIndex DataFrames` or the `xarray` library for three-dimensional data. If you're working with the latest versions of Pandas, you may not encounter the `Panel` data structure, and it's not a common choice for handling data compared to `Series` and `DataFrame`.

```python

series1 = pd.Series([1, 2, 3])
series2 = pd.Series(['apple', 'banana', 'orange'])
series3 = pd.Series([10.5, 20.3, 15.8])

df3 = pd.DataFrame({'A':series1,'B':series2,'C':series3})

print(df3)
```
