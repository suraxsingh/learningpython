```python
#Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.
#ZeroDivisionError and OverflowError are defined in the ArithmeticError class.
#ZeroDivisionError
try:
    result = 10 / 0
except ZeroDivisionError as e:
    print("Error:",e)
#OverflowError
import sys

try:
    result = sys.maxsize * sys.maxsize
except OverflowError:
    print("Error: The result of the operation is too large to be represented.")
```

```python
#Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.
# The LookupError class in Python is used as a base class for exceptions that
# occur when a key or index is not found during a lookup operation.
# It provides a common superclass for specific lookup-related exceptions,
# such as KeyError and IndexError.
# KeyError
my_dict = {"apple": 3, "banana": 6, "orange": 2}

try:
    count = my_dict["grape"]
except KeyError:
    print("Error: The key does not exist in the dictionary.")

#IndexError
my_list = [1, 2, 3, 4, 5]

try:
    value = my_list[10]
except IndexError:
    print("Error: The index is out of range.")
```

#Q5. Explain ImportError. What is ModuleNotFoundError?
ImportError is an exception class in Python that is raised when an import statement fails to locate and import a module or when there is a problem with the imported module itself.


The distinction between ImportError and ModuleNotFoundError lies in their specificity. While ImportError is a broader exception that covers import-related errors, ModuleNotFoundError is a more specialized exception that specifically indicates that a module could not be found during import. By using ModuleNotFoundError, you can handle cases where the failure to import is due to the absence of the requested module with more clarity.

#Q6. List down some best practices for exception handling in python.


1. Catch specific exceptions rather than using a generic `except` block.
2. Use multiple `except` blocks for different exception types.
3. Utilize the `finally` block for cleanup tasks.
4. Avoid catching and ignoring exceptions without action.
5. Wrap specific lines of code with `try-except` blocks.
6. Provide meaningful error messages.
7. Avoid bare `except` statements without specifying exception types.
8. Consider logging exceptions for better error management.
9. Use context managers, like the `with` statement, for resource management.
10. Document your exception handling strategy and the expected exceptions.
